```python
# NL Fix

NL1 = (1.305 + 1.340 + 1.137) / 3
NL2 = (1.527 + 1.239 + 1.205) / 3
NL3 = (1.339 + 1.132 + 1.285 + 1.187) / 4
NL4 = (1.170 + 1.225) / 2

nlfix = {
    "NL1": NL1,
    "NL2": NL2,
    "NL3": NL3,
    "NL4": NL4,
}

for reg, val in nlfix.items():
    qog_final.loc[qog_final['Region_ID'] == reg, 'QoG'] = val
```
