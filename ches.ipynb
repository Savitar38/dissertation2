## Partyfacts integration

```python
pf = pd.read_csv("./datasets/partyfacts-external-parties.csv")
```

```python
pf = pf[pf["dataset_key"] == "ches"][["country", "dataset_party_id", "name_short", "partyfacts_id"]]
```

```python
pf
```

```python
pf["dataset_party_id"] = pd.to_numeric(pf["dataset_party_id"])
```

```python
ches = pd.merge(df, pf, left_on="party_id", right_on="dataset_party_id", how="left")
```

```python
ches[ches["partyfacts_id"].isnull()]
```

```python
ches = ches[["partyfacts_id", "year", "country_y", "name_short", "vote", "electionyear", "govt", "family", "eu_position", "lrgen", "lrecon", "people_vs_elite"]]
```

```python
ches = ches.rename(columns={"year" : "year_ches", "country_y" : "country_ches", "name_short" : "party_name_short", "vote" : "vote_perc_ches"})
```

```python
ches.to_csv("./datasets/clean/ches-clean.csv", index=False)
```
