### Cleaning the dataset

```python
df["Fund"].unique()
```

```python
df["TO"].unique()
```

```python
df["ms"].unique() # TC == Interreg
```

```python
df[(df["TO"] == "DM")]
```

- as per Santos, removing TO 13 (fostering crisis repair and resilience), TO 12 (outermost and
sparsely populated), technical assistance programmes or axes, and discontinued programmes, due to being specific to geography
- "Programmes that are discontinued over time because they were merged with others are also left out from the analysis." HOW TO FIND THESE? -> they include "merged" in their title
- also remove UK due to Brexit and following complications

```python
df = df[~(df["TO"].isin(["13", "12", "TA", "IPA-TA"])) & ~(df["ms"].isin(["TC", "UK"]))]
```

```python
df = df[~(df["Programme Title"].str.contains("merge", na=False, case=False))]
```

- "although this analysis covers the programing period 2014–2020, it considers 2016 as the first year of implementation given delays in launching the programmes"

```python
df = df[df["year"] >= 2016]
```

```python
df_s = df[["MS_name", "year","Fund", "Total_Amount_planned", "total_eligible_spending"]]

df_s = df_s[((df_s["Fund"] == "ESF") | (df_s["Fund"] == "ERDF") | (df_s["Fund"] == "CF")) & (df_s["year"] == 2020)]

df_s["Total_Amount_planned"] = df_s["Total_Amount_planned"].str.replace(",", "")
df_s["Total_Amount_planned"] = df_s["Total_Amount_planned"].astype(float)

df_s["total_eligible_spending"] = df_s["total_eligible_spending"].str.replace(",", "")
df_s["total_eligible_spending"] = df_s["total_eligible_spending"].astype(float)
```

```python
df_s
```

```python
totals = df_s.groupby('MS_name').agg({
    'Total_Amount_planned': 'sum',
    'total_eligible_spending': 'sum'
}).reset_index()

totals['Fund'] = 'TOTAL'

totals['rate'] = (totals['total_eligible_spending'] / totals['Total_Amount_planned']) * 100

totals
```

```python
plt.title("Absorption rates TOTAL in 2014-2020")
plt.barh(totals.sort_values(by="rate")["MS_name"], totals.sort_values(by="rate")["rate"])
plt.show()
```
