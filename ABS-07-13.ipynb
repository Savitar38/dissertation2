```python
import pandas as pd
import matplotlib.pyplot as plt
```

https://cohesiondata.ec.europa.eu/2007-2013-Finances/SF-2007-2013-Funds-Absorption-Rate/kk86-ceun/about_data

```python
df = pd.read_csv("ABS-07-13.csv")
```

```python
df["Fund"].unique()
```

```python
# SVK sorted by year

df[df["Member State"] == "Slovakia"].sort_values(by="Year")
```

```python
# SVK total sorter by year

df[(df["Member State"] == "Slovakia") & (df["Fund"] == "TOTAL")].sort_values(by="Year")
```

```python
# All, 2010, total, by abs%
df_10 = df[(df["Year"] == 2013) & (df["Fund"] == "TOTAL")].sort_values(by="Absorption (%)")

# All, 2010, ESF, by abs%
df_10_esf = df[(df["Year"] == 2013) & (df["Fund"] == "ESF")].sort_values(by="Absorption (%)")

# All, 2010, ERDF, by abs%
df_10_erdf = df[(df["Year"] == 2013) & (df["Fund"] == "ERDF")].sort_values(by="Absorption (%)")

# All, 2010, CF, by abs%
df_10_cf = df[(df["Year"] == 2013) & (df["Fund"] == "CF")].sort_values(by="Absorption (%)")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 18))

ax1 = axes[0, 0]
ax1.set_title('Fund: TOTAL')
ax1.barh(df_10["Member State"], df_10["Absorption (%)"], height=0.5)

ax2 = axes[0, 1]
ax2.set_title('Fund: ESF')
ax2.barh(df_10_esf["Member State"], df_10_esf["Absorption (%)"], height=0.5)

ax3 = axes[1, 0]
ax3.set_title('Fund: ERDF')
ax3.barh(df_10_erdf["Member State"], df_10_erdf["Absorption (%)"], height=0.5)

ax4 = axes[1, 1]
ax4.set_title('Fund: CF')
ax4.barh(df_10_cf["Member State"], df_10_cf["Absorption (%)"], height=0.5)
```

```python
# All, 2015, total, by abs%
df_15 = df[(df["Year"] == 2015) & (df["Fund"] == "TOTAL")].sort_values(by="Absorption (%)")

# All, 2015, ESF, by abs%
df_15_esf = df[(df["Year"] == 2015) & (df["Fund"] == "ESF")].sort_values(by="Absorption (%)")

# All, 2015, ERDF, by abs%
df_15_erdf = df[(df["Year"] == 2015) & (df["Fund"] == "ERDF")].sort_values(by="Absorption (%)")

# All, 2015, CF, by abs%
df_15_cf = df[(df["Year"] == 2015) & (df["Fund"] == "CF")].sort_values(by="Absorption (%)")
```

```python
fig2, axes2 = plt.subplots(2, 2, figsize=(12, 18))

ax5 = axes2[0, 0]
ax5.set_title('Fund: TOTAL')
ax5.barh(df_15["Member State"], df_15["Absorption (%)"], height=0.5)

ax6 = axes2[0, 1]
ax6.set_title('Fund: ESF')
ax6.barh(df_15_esf["Member State"], df_15_esf["Absorption (%)"], height=0.5)

ax7 = axes2[1, 0]
ax7.set_title('Fund: ERDF')
ax7.barh(df_15_erdf["Member State"], df_15_erdf["Absorption (%)"], height=0.5)

ax8 = axes2[1, 1]
ax8.set_title('Fund: CF')
ax8.barh(df_15_cf["Member State"], df_15_cf["Absorption (%)"], height=0.5)
```
